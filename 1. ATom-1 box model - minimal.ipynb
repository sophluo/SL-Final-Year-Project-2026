## Part 2: Constructing the box model

The steady-state approximation used is:

$[OH]_{ss}=\frac{(2k_1[O(^1D)][H_2O])}{(k_2 [CO]+k_3 [CH_4])}$

Or, put more simply: 

$[OH]_{ss}=\frac{source \: rate}{sink \: rate \: coefficient}$

There are no direct ATom measurements for $[O^1D]$, however because it is short-lived, similar to $OH$ we can calculate a steady-state value for $[O^1D]$ using $[O^1D]_{SS} = \frac{J[O_3]}{k_1[H_2O] + k_5[M]}$

(After the conversion to $molecules \: cm^{-3}$ for the other concentrations, $[M]$ cancels out to become 1, but it is technically still included in the rate equation)

Source:

$O_3 + J → O_2 + O(^1D)$

Sink:

$O(^1D) + M → O(^3P) + M$

$O(^1D) + H_2O → 2OH$ (this is the same as the source rate of $OH$)

The calculations will be carried out in the following order:
- Calculation of the source rate constants ($k_1$, $k_2$, $k_3$, $k_4$)
- Calculation of the source rate
- Calculation of the sink rate coefficients
- Calculation of $[OH]$

## Calculation of the rate constants
Temperature and pressure are variable at different altitudes in the troposphere so we want to use Arrhenius equation to calculate the rate constants: $k = Ae^{-\frac{E_a}{RT}}$

(the exception is $k_{OH+CO}$ which follows non-Arrhenius behaviour and so a different approximation is used, see report for details on this)

```python
df['kO1DH2O'] = 1.63e-10 * np.exp(-(-60/df["Temp"])) ## page 21 of https://jpldataeval.jpl.nasa.gov/pdf/NASA-JPL%20Evaluation%2019-5.pdf
df['kO1DM'] = 2.15e-11 * np.exp(-(-110/df["Temp"])) ## page 21 of https://jpldataeval.jpl.nasa.gov/pdf/NASA-JPL%20Evaluation%2019-5.pdf
```

### Check that the source rate constants at 298 K match with literature values

```python
print("k1 for reaction of O1D + H2O:", 1.63e-10 * np.exp(-(-60/298)), "lit: 2.0e-10")
print("k5 for reaction of O1D + M:", 2.15e-11 * np.exp(-(-110/298)), "lit: 3.1e-11")
```

Before proceeding to calculate the source rate we need to convert units for the species so that everything is in units of $molecules \: cm^{-3}$ and calculate $[M]$ using $[M] = \frac{p}{kT}$ ($p$ is pressure in $Pa$, $k$ is Boltzmann's constant in $J \: K^{-1}$ and $T$ is temperature in $K$

$1 J = 1 \: kg \: m^2 \: s^{-2}$

$1 \: Pa = 1 \: N m^{-2}$

Multiply value of $[M]$ by $1 \times 10^{-6}$, so from $molecules \: m^{-3}$ to $molecules \: cm^{-3}$

```python
kb = 1.38e-23 # J K^−1 Boltzmann's constant

for i in range(len(df)):
    df.loc[i, "M"] = ((df.loc[i, "Pres"] * 100) / (kb * df.loc[i, "Temp"])) * 1e-6 ## molecules cm^3, pressure was in hPa so multiplied by 100 to Pa

df.head()
```

## Source rate

The source rate of $OH$ is $2k[O(^1D)][H_2O]$



```python
# New columns for conversion from mole fraction to molecules cm^-3
df["O3"] = df["O3_M"] * df["M"]
df["H2O"] = df["H2O_M"] * df["M"]
df["HO2"] = df["HO2_ATHOS"] * 1e-12 * df["M"]
df["NO"] = df["NOX_M"] * df["M"]

df["O1D"] = (df["jO3_O2_O1D_CAFS"] * df["O3_M"]) / ((df['kO1DH2O'] * df["H2O_M"]) + (df['kO1DM']))
```

```python
for i in range(len(df)):
    df.loc[i, "source"] = (2 * df['kO1DH2O'][i] * df["O1D"][i] * df["H2O"][i]) 
print(f"Mean source rate: {df["source"].mean():.2e}")
df.head()
# source should be around 1e6 molecules cm^-3 s^-1

print("Source rates (in molecules cm^-3 s^-1):")
print(f"Mean O1D contribution:  {(2 * df['kO1DH2O'] * df["O1D"] * df["H2O"]).mean():.2e}") # This is currently the only source, so should be same value
```

## Finding the sink rate coefficients

Use the sink rate coefficients here

$OH + CO → H + CO2$

Rate coefficient = $k[CO]$

$OH + CH_4 → CH_3 + H_2O$

Rate coefficient = $k[CH_4]$

```python
# Calculating rate coefficients for sink terms
df["P_atm"] = 0.000987 * df["Pres"]
df['kOHCO']  = 1.5e-13  * (1.0 + 0.6 * df["P_atm"]) #where P is the pressure in atmospheres
df['kOHCH4'] = 2.45e-12 * np.exp(-(1775/df['Temp'])) ## page 106 of https://jpldataeval.jpl.nasa.gov/pdf/NASA-JPL%20Evaluation%2019-5.pdf

df["CO"] = df['CO_M'] * df["M"]
df["CH4"] = df['CH4_M'] * df["M"]

# Sink contributions
print("Sink rate coefficients (in molecules^-1 cm^3 s^-1):")
print(f"CO contribution: {round((df['kOHCO'] * df["CO"]).mean(), 3)}")
print(f"CH4 contribution: {round((df['kOHCH4'] * df["CH4"]).mean(), 3)}")
```

### Check that the sink rate constants at 298 K match with literature values

```python
# print("k for reaction of OH + CO:", 1.85e-13 * np.exp(-(65/298)), "lit: 2.4e-13")
print(f"k for reaction of OH + CO: {(1.5e-13  * (1.0 + 0.6 * 1)):.2e}, lit: 2.4e-13")
print(f"k for reaction of OH + CH4: {(2.45e-12 * np.exp(-(1775/298))):.2e}, lit: 6.3e-15")
```

## Calculating $[OH]$

```python
# Calculating [OH]
for i in range(len(df)):
    df.loc[i, "[OH]_calc"] = (df.loc[i, "source"]) / ((df['kOHCO'][i] * df["CO"][i]) + (df['kOHCH4'][i] * df["CH4"][i]))
```

We can also see that box model tends to underpredict $[OH]$ at high altitudes for these two dates. For a more quantitative assessment of this, can calculate some metrics (i.e. $r^2$ and RMSE).

```python
X = df[["[OH]_measured"]][(df["date"] > "2016-07-28") & (df["date"] < "2016-07-30")]
y = df["[OH]_calc"][(df["date"] > "2016-07-28") & (df["date"] < "2016-08-01")]
print(f"r2 for 28/07/2016: {round(sklearn.metrics.r2_score(X, y), 3)}")
print(f"RMSE for 28/07/2016: {sklearn.metrics.root_mean_squared_error(X, y):.2e} molecules cm^-3")

X = df[["[OH]_measured"]][(df["date"] > "2016-08-05") & (df["date"] < "2016-08-07")]
y = df["[OH]_calc"][(df["date"] > "2016-08-05") & (df["date"] < "2016-08-07")]
print(f"r2 for 06/08/2016: {round(sklearn.metrics.r2_score(X, y), 3)}")
print(f"RMSE for 06/08/2016: {sklearn.metrics.root_mean_squared_error(X, y):.2e} molecules cm^-3")

X = df[["[OH]_measured"]]
y = df["[OH]_calc"]
print(f"r2 for whole dataset: {round(sklearn.metrics.r2_score(X, y), 4)}")
print(f"RMSE for whole dataset: {sklearn.metrics.root_mean_squared_error(X, y):.2e} molecules cm^-3")
```

$r^2$ for whole dataset is higher than the $r^2$ for the individual dates - likely due to smaller sample sizes when filtering for individual dates meaning any outliers result in a larger unexplained variance leading to lower $r^2$.

Can see if there is trend between $r^2$ or RMSE and altitude

First, take a look at distributions of altitudes across the dataset:

There are more data points at high altitudes, so at say above 6000 m, can take averages at more frequent intervals.

```python
numbers = list(range(0, 6000, 1000))
numbers = numbers + (list(range(6000, 13000, 500)))
metrics_r2 = []
metrics_rmse = []
for i in range(len(numbers) - 1):
    # print(numbers[i+1])
    X = df[["[OH]_measured"]][(df["Altitude"] >= int(numbers[i])) & (df["Altitude"] < int(numbers[i + 1]))]
    y = df["[OH]_calc"][(df["Altitude"] >= int(numbers[i])) & (df["Altitude"] < int(numbers[i + 1]))]
    metrics_r2.append(sklearn.metrics.r2_score(X, y))
    metrics_rmse.append(sklearn.metrics.root_mean_squared_error(X, y))
```

```python
fig, ax = plt.subplots(figsize = (8, 6))
plt.rcParams.update({'font.size':15})

plt.scatter(numbers[:-1], metrics_r2)
plt.axhline(y=1, color='r', linestyle='--', label='${r^2}$ = 1, perfect fit')
plt.axhline(y=0, color='b', linestyle='--', label='${r^2}$ = 0, imperfect fit')
plt.axhline(y = (sklearn.metrics.r2_score(df[["[OH]_measured"]], df["[OH]_calc"])), color = 'g', linestyle = '--', label = f'$r^2$ = {round(sklearn.metrics.r2_score(df[["[OH]_measured"]], df["[OH]_calc"]), 4)}, average across dataset')
plt.xlabel("Altitude / m")
plt.ylabel("Average $r^2$ value")
plt.title("$r^2$ values at different intervals of altitudes")
plt.legend(prop={'size':15})

plt.grid()
```

```python
fig, ax = plt.subplots(figsize = (8, 6))
plt.rcParams.update({'font.size':15})

plt.scatter(numbers[:-1], metrics_rmse)
plt.axhline(y = (sklearn.metrics.root_mean_squared_error(df[["[OH]_measured"]], df["[OH]_calc"])), color = 'g', linestyle = '--', label = f'RMSE = {'{:.2e}'.format(sklearn.metrics.root_mean_squared_error(df[["[OH]_measured"]], df["[OH]_calc"]))}, average across dataset')
plt.xlabel("Altitude / m")
plt.ylabel("Average RMSE value")
plt.title("RMSE values at different intervals of altitudes")
plt.legend(prop={'size':15})

plt.grid()
```

Both plots show poorer agreement/higher residuals at low altitudes of 2000 m to 4000 m and high altitudes of above 10000 m. It is likely that this is due to missing source/sink reactions, or processes not accounted for in the model, e.g. emission/deposition.

## Summary

This box model with a minimal set of reactions showed the best agreement at middle altitudes of 4000 m-6000 m, however struggled at low altitudes and high altitudes. The disagreement at high altitudes was due to colder temperatures leading to other source processes being more dominant. The disagreement at low altitudes was likely due to deposition and emission processes not included in the model, as well as chemistry that takes place in the lower troposphere such as reactions with VOCs.

The model assumes that $O(^1D)+H_2O$ was the only source of $OH$. However, in reality, there are a cycle of reactions controlling [OH] and so it may be necessary to consider these additional reactions that regenerate $OH$. This is considered in the Jupyter notebook `2. ATom-1 box model - with HOx and NOx reactions`
