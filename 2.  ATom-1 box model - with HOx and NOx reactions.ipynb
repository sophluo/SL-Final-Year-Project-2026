# 2.  ATom-1 box model - with HOx and NOx reactions

This notebook contains calculations for the zero-dimensional box model with an extended set of inputs.

It follows the same procedure as the calculations with the minimal set of inputs, but with additional source and sink reactions.

These additional reactions are:

**Sources:**

**$HO_2 + NO → OH + NO_2$**

rate = $k_4[HO_2][NO]$

**$HO_2 + O_3→OH + 2O_2$**

rate = $k_5[HO_2][O_3]$

**Sinks:**

**$OH + O_3 → HO_2 + O_2$**

rate coefficient = $k_6[O_3]$

**$OH + HNO_3 → H_2O + NO_2$**

rate coefficient = $k_7[HNO_3]$

**$OH + HO_2 → H_2O + O_2$**

rate coefficient = $k_8[HO_2]$

The updated equation for $[OH]$ is now:

$[OH]_{ss}=\frac{2k_1[O(^1D)][H_2O] + k_5[HO_2][NO] + k_6[HO_2][O_3]}{k_2[CO] + k_3[CH_4] + k_7[O_3] + k_8[HO_2] + k_9[HNO_3]}$

---

```python
# Importing libraries
import pandas as pd
from datetime import datetime, timezone
import numpy as np
from pytz import timezone
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pytz
import sklearn
from sklearn import metrics
# If cartopy is not installed, run this
# !pip install cartopy
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
```

```python
# Read csv file as a pandas dataframe
df = pd.read_csv("MDS_atom1_2016_summer_with_no_no2_oh_ho2_full_js.csv")
```

```python
# Filtering to only keep columns that are used, and output a few rows
df = df[['Temp', 'Pres','UTC_Start_dt', 'lat', 'lon', 'Altitude', 'jO3_O2_O1D_CAFS', 'O3_M', 'H2O_M', 'OH_ATHOS', 'HO2_ATHOS', 'CO_M', 'CH4_M', 'PAN_M', 'NOX_M', 'NO_CL', 'HNO3_M']]
df.head()
```

```python
# Converting timezone to New York as this is a more representative timezone
df["UTC_Start_dt"] = pd.to_datetime(df["UTC_Start_dt"])
df['UTC_Start_dt'] = df['UTC_Start_dt'].dt.tz_localize('UTC')
df["UTC_Start_dt"] = df["UTC_Start_dt"].dt.tz_convert('America/New_York')
df['date'] = df['UTC_Start_dt']
df['date']
```

```python
# For unit conversions from mole fraction to molecules cm^-3, calculate M
kb = 1.38e-23  # J K^−1 Boltzmann's constant

for i in range(len(df)):
    df.loc[i, "M"] = ((df.loc[i, "Pres"] * 100) / (kb * df.loc[i, "Temp"])) * 1e-6  # molecules cm^3, pressure was in hPa so multiplied by 100 to Pa

df.head()
```

## Source rate

**Additional sources being added:**

**$HO_2 + NO → OH + NO_2$**

**source rate = $k_4[HO_2][NO]$**

**$HO_2 + O_3→OH + 2O_2$**

**source rate = $k_5[HO_2][O_3]$**

This now makes the source rate:


$2k_1[O(^1D)][H_2O]) + k_4[HO_2][NO] + k_5[HO_2][O_3]$

```python
# Same calculations that were done for minimal box model
df['kH2O'] = 1.63e-10 * np.exp(-(-60/df["Temp"])) ## k1, page 21 of https://jpldataeval.jpl.nasa.gov/pdf/NASA-JPL%20Evaluation%2019-5.pdf
df['kM'] = 2.15e-11 * np.exp(-(-110/df["Temp"])) ## k5, page 21 of https://jpldataeval.jpl.nasa.gov/pdf/NASA-JPL%20Evaluation%2019-5.pdf

# Calculation of rate constants for additional sources
df['kHO2NO'] = 3.44e-12 * np.exp(-(-260/df["Temp"])) ## k4, page 83 of https://jpldataeval.jpl.nasa.gov/pdf/NASA-JPL%20Evaluation%2019-5.pdf
df['kHO2O3'] = 1.0e-14 * np.exp(-(490/df["Temp"])) ## k5, page 67 of https://jpldataeval.jpl.nasa.gov/pdf/NASA-JPL%20Evaluation%2019-5.pdf

# New column for conversion from mole fraction to molecules cm^-3
df["O3"] = df["O3_M"] * df["M"]
df["H2O"] = df["H2O_M"] * df["M"]
df["HO2"] = df["HO2_ATHOS"] * 1e-12 * df["M"]
df["NO"] = df["NO_CL"] * 1e-9 * df["M"]

# Steady-state approximation of [O1D]
df["O1D"] = (df["jO3_O2_O1D_CAFS"] * df["O3_M"]) / ((df['kH2O'] * df["H2O_M"]) + (df['kM']))
```

```python
for i in range(len(df)):
    df.loc[i, "source"] = (2 * df['kH2O'][i] * df["O1D"][i] * df["H2O"][i]) + (df['kHO2O3'][i] * df["HO2"][i] * df["O3"][i]) + (df['kHO2O3'][i] * df["HO2"][i] * df["O3"][i])+ (df['kHO2NO'][i] * df["HO2"][i] * df["NO"][i])  ## Error
print(f"Mean source rate: {df["source"].mean():.2e}") # source should be around 1e6 molecules cm^-3 s^-1

print("Mean contribution to source rates (in molecules cm^-3 s^-1):")
print(f"Mean O1D contribution:  {(2 * df['kH2O'] * df["O1D"] * df["H2O"]).mean():.2e}")
print(f"Mean NO contribution: {(df['kHO2NO'] * df["HO2"] * df["NO"]).mean():.2e}")
print(f"HO2 contribution: {(df['kHO2O3'] * df["HO2"] * df["O3"]).mean():.2e}")
```

### Check that the source rate constants at 298 K match with literature values

```python
print(f"k1 for reaction of O1D + H2O: {(1.63e-10 * np.exp(-(-60/298))):.2e} cm^3 molecules ^-1 s^-1, lit: 2.0e-10")
print(f"k4 for reaction of HO2 + NO: {(3.44e-12 * np.exp(-(-260/298))):.2e} cm^3 molecules ^-1 s^-1, lit: 8.2e-12")
print(f"k5 for reaction of HO2 + O3: {(1.0e-14 * np.exp(-(490/298))):.2e} cm^3 molecules ^-1 s^-1, lit: 1.9e-15")
```

## Sink rate coefficient

**Additional sinks being added:**

**$OH + O_3 → HO_2 + O_2$**

**rate coefficient $k_6[O_3]$**

**$OH + HNO_3 → H_2O + NO_2$**

**rate coefficient $k_7[HNO_3]$**

**$OH + HO_2 → H_2O + O_2$**

**rate coefficient $k_8[HO_2]$**

This makes the sink rate coefficient: 

$k_2[CO] + k_3[CH_4] + k_7[O_3] + k_8[HO_2] + k_9[HNO_3]$

```python
df["P_atm"] = 0.000987 * df["Pres"] # Calculating P in atmospheres, to use for approximation of Troe formula

# The same rate constants used for minimal box model
df['kOHCO']  = 1.5e-13  * (1.0 + 0.6 * df["P_atm"]) # k2, where P is the pressure in atmospheres, page 484 of https://jpldataeval.jpl.nasa.gov/pdf/NASA-JPL%20Evaluation%2019-5.pdf
df['kOHCH4'] = 2.45e-12 * np.exp(-(1775/df['Temp'])) ## k3, page 106 of https://jpldataeval.jpl.nasa.gov/pdf/NASA-JPL%20Evaluation%2019-5.pdf

# Additional rate constants being calculated
df['kOHO3'] = 1.7e-12 * np.exp(-(940/df["Temp"])) ## k7, page 67 of https://jpldataeval.jpl.nasa.gov/pdf/NASA-JPL%20Evaluation%2019-5.pdf
df['kOHHO2'] = 4.8e-11 * np.exp(-(-250/df["Temp"])) ## k8, page 67 of https://jpldataeval.jpl.nasa.gov/pdf/NASA-JPL%20Evaluation%2019-5.pdf
df['kOHHNO3'] = 8.28e-14 * (1.0 + 0.6 * df["P_atm"]) ## k9, page 484 of https://jpldataeval.jpl.nasa.gov/pdf/NASA-JPL%20Evaluation%2019-5.pdf

df["CO"] = df['CO_M'] * df["M"]
df["CH4"] = df['CH4_M'] * df["M"]
df["HNO3"] = df['HNO3_M'] * df["M"] # New species being used for this box model

# Sink contributions
print("Sink rate coefficients (in molecules^-1 cm^3 s^-1):")
print(f"CO contribution: {(df['kOHCO'] * df["CO"]).mean():.2e}")
print(f"CH4 contribution: {(df['kOHCH4'] * df["CH4"]).mean():.2e}")
print(f"O3 contribution: {(df['kOHO3'] * df["O3"]).mean():.2e}")
print(f"HO2 contribution: {(df['kOHHO2'] * df["HO2"]).mean():.2e}")
print(f"HNO3 contribution: {(df['kOHHNO3'][i] * df["HNO3"][i]).mean():.2e}")
```

### Check that the sink rate constants at 298 K match with literature values

```python
print(f"k2 for reaction of OH + CO: {(1.5e-13  * (1.0 + 0.6 * 1)):.2e}, lit: 2.4e-13")
print(f"k3 for reaction of OH + CH4: {(2.45e-12 * np.exp(-(1775/298))):.2e}, lit: 6.3e-15")
print(f"k7 for reaction of OH + O3: {(1.7e-12 * np.exp(-(940/298))):.2e}, lit: 7.3e-14")
print(f"k8 for reaction of OH + HO2: {(4.8e-11 * np.exp(-(-250/298))):.2e}, lit: 1.1e-10")
print(f"k9 for reaction of OH + HNO3: {(8.28e-14  * (1.0 + 0.6 * 1)):.2e}, lit: 1.4e-13")
```

## Finally, calculating $[OH]$

```python
# Calculating [OH]
for i in range(len(df)):
    df.loc[i, "[OH]_calc"] = (df.loc[i, "source"]) / ((df['kOHCO'][i] * df["CO"][i]) + (df['kOHCH4'][i] * df["CH4"][i]) + (df['kOHHNO3'][i] * df["HNO3"][i]) + (df['kOHHO2'][i] * df["HO2"][i]) + (df['kOHO3'][i] * df["O3"][i]))
```

Again, calculate $r^2$ and RMSE at these different altitudes to see if there is a similar trend observed from the minimal box model.

```python
numbers = list(range(0, 6000, 1000))
numbers = numbers + (list(range(6000, 13000, 500)))
metrics_r2 = []
metrics_rmse = []
for i in range(len(numbers) - 1):
    X = df[["[OH]_measured"]][(df["Altitude"] >= int(numbers[i])) & (df["Altitude"] < int(numbers[i + 1]))]
    y = df["[OH]_calc"][(df["Altitude"] >= int(numbers[i])) & (df["Altitude"] < int(numbers[i + 1]))]
    metrics_r2.append(sklearn.metrics.r2_score(X, y))
    metrics_rmse.append(sklearn.metrics.root_mean_squared_error(X, y))
```

```python
fig, ax = plt.subplots(figsize = (8, 6))
plt.rcParams.update({'font.size':15})

plt.scatter(numbers[:-1], metrics_r2)
plt.axhline(y=1, color='r', linestyle='--', label='${r^2}$ = 1, perfect fit')
plt.axhline(y=0, color='b', linestyle='--', label='${r^2}$ = 0, imperfect fit')
plt.axhline(y = (sklearn.metrics.r2_score(df[["[OH]_measured"]], df["[OH]_calc"])), color = 'g', linestyle = '--', label = f'$r^2$ = {round(sklearn.metrics.r2_score(df[["[OH]_measured"]], df["[OH]_calc"]), 4)}, average across dataset')
plt.xlabel("Altitude / m")
plt.ylabel("A$r^2$ value")
plt.title("$r^2$ values at different intervals of altitudes")
plt.legend(prop={'size':15})

plt.grid()
```

```python
fig, ax = plt.subplots(figsize = (8, 6))
plt.rcParams.update({'font.size':15})

plt.scatter(numbers[:-1], metrics_rmse)
plt.axhline(y = (sklearn.metrics.root_mean_squared_error(df[["[OH]_measured"]], df["[OH]_calc"])), color = 'g', linestyle = '--', label =  f'RMSE = {'{:.2e}'.format(sklearn.metrics.root_mean_squared_error(df[["[OH]_measured"]], df["[OH]_calc"]))}, average across dataset')
plt.xlabel("Altitude / m")
plt.ylabel("RMSE value / molecules cm$^{-3}$")
plt.title("RMSE values at different intervals of altitudes")
plt.legend(prop={'size':15})

plt.grid()
```

### Plot to identify lat/lon of high RMSE at the low altitudes of below 2000 m

```python
df["residual"] = np.nan
for i in range(len(df)):
    df.loc[i, "residual"] = abs(df.loc[i, "[OH]_measured"] - df.loc[i, "[OH]_calc"])
mask = (df["Altitude"] <= 2000) & (df["Altitude"] >= 0) & (df["residual"] > sklearn.metrics.root_mean_squared_error(df["[OH]_measured"], df["[OH]_calc"]).mean())

ax = plt.axes(projection=ccrs.PlateCarree(central_longitude=240))
ax.stock_img()
plt.rcParams.update({'font.size':10})
plt.plot(df["lon"], df["lat"], transform=ccrs.Geodetic())
plt.scatter(df["lon"][mask], df["lat"][mask], label="above average RMSE at altitude of <= 2000 m", transform=ccrs.Geodetic())
ax.set_xlabel("longitude")
ax.set_ylabel("latitude")
plt.title("ATom 1 flight trajectory")
plt.legend()
plt.show()
```

### Summary
The disagreement at high altitudes was due to colder temperatures leading to other source processes being more dominant, and this was shown by an improvement when including $NO_x$ and $HO_x$ interconversion reactions. The disagreement at low altitudes was likely due to deposition and emission processes not included in the model, as well as chemistry that takes place in the lower troposphere such as reactions with VOCs.Surprisingly, the overall $r^2$ score was lower and negative for this updated box model (-0.173 compared to 0.0958) and was shown to originate from poorer agreement at low altitudes.
